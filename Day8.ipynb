```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_-_FY2020.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

### -scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.

```python
x=df['Total Savings']
x
```

```python
y=df['Job created']
y
```

```python
df.groupby(['Total Savings','Job created']).mean()['Savings from beginning receiving benefits']
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.scatter(x,y,color='red')
plt.xlabel('Total Savings')
plt.ylabel('Job created')
plt.title('Standard Scale')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.scatter(x,y,color='black',marker='*')
plt.xscale("log")
plt.xlabel('Total Savings')
plt.ylabel('Job created')
plt.title('Logaritmic Scale')
plt.show()
```

### -histogram of the log of the average total savings.

```python
x1=df['Total Savings']
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.hist(x1,log=True,bins=25)
plt.xlabel('Total Savings')
plt.title('Histogram')
plt.show()
```

### -line plot of the total jobs created for each month.¶

```python
df_jobcreated=df[~df['Job created'].isnull()]
df_jobcreated
```

```python
df_jobcreated['Effective Date']
```

```python
df_jobcreated['Effective Date'] = pd.to_datetime(df['Effective Date'])
print (df_jobcreated['Effective Date'])
```

```python
x3=df_jobcreated['Effective Date'].sort_values()
x3
```

```python
y3=df_jobcreated['Job created']
y3
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x3,y3,color='purple',marker='D')
plt.xlabel('Effective Date')
plt.ylabel('Job created')
plt.title('Line Plot')
plt.show()
```
